```python
## Join the Dataframe based on Common Fields
```

```python
# First, merge sales with customers on 'Customer Number'
sales_customers = pd.merge(sales, customers, on='Customer Number', how='left')
print (sales_customers)
```

```python
# Next, merge the result with cities on 'City Code' to get regional data
sales_customers_cities = pd.merge(sales_customers, cities, on='City Code', how='left')
print (sales_customers_cities)
```

```python
# Merge with the item_master on 'Item Number' to get product details
sales_full = pd.merge(sales_customers_cities, item_master, on='Item Number', how='left')
print (sales_full)
sales_full
```

```python
# Finally, merge with sales_rep on 'Sales Rep Number' to include sales representative details
sales_full = pd.merge(sales_full, sales_rep, left_on='Sales Rep Number', right_on='Sales Rep ID', how='left')
print (sales_full)
sales_full
```

```python
# Analyze Total Sales by Sales Rep

# Step 1: Combine the sales rep columns into one column
sales_reps_combined = pd.melt(
    sales_full[['Sales Rep Name_y', 'Sales Rep Name1_y', 'Sales Rep Name2_y', 'Sales Rep Name3_y', 'Sales']],
    id_vars='Sales',  # Keep the 'Sales' column intact
    value_name='Sales Rep'  # Name the new combined column 'Sales Rep'
).dropna()  # Remove rows with missing sales reps

# Step 2: Group by 'Sales Rep' and sum up the total sales for each rep
sales_by_rep_combined = sales_reps_combined.groupby('Sales Rep')['Sales'].sum().sort_values(ascending=False)

# Step 3: Convert sales to millions USD
sales_by_rep_combined_millions = sales_by_rep_combined / 1e6

# Step 4: Rank the sales reps by total sales
sales_by_rep_ranked = sales_by_rep_combined_millions.rank(ascending=False)

# Display the ranked sales reps
print(sales_by_rep_ranked)

# Step 5: Plot the sales rep totals in millions USD
sales_by_rep_combined_millions.plot(kind="bar", figsize=(20,8), color='skyblue')
plt.title('Total Sales by Sales Rep (in Millions US$)')
plt.xlabel('Sales Rep')
plt.ylabel('Total Sales (Millions US$)')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.show()
```

```python
# Analyze Sales by Product Types

# Step 3: Sales by Product Group and Product Type
sales_by_product_group = sales_full.groupby('Product Group')['Sales'].sum().sort_values(ascending=False)
sales_by_product_type = sales_full.groupby('Product Type')['Sales'].sum().sort_values(ascending=False)

# Convert sales to millions USD
sales_by_product_type_millions = sales_by_product_type / 1e6

# Plotting the sales by product type in millions of USD
sales_by_product_type_millions.plot(kind="bar", figsize=(16,6), color='lightgreen')
plt.title('Total Sales by Product Type (in Millions US$)')
plt.xlabel('Product Type')
plt.ylabel('Total Sales (Millions US$)')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.show()
```

```python
# Analyze Supply Chain Logistics
# Converting the 'Promised Delivery Date' and 'Date' columns to datetime for analysis

sales_full ['Promised Delivery Date'] = pd.to_datetime(sales_full ['Promised Delivery Date'], errors='coerce')
sales_full ['Date'] = pd.to_datetime(sales_full ['Date'], errors='coerce')

# Create a new column to calculate delivery delay in days
sales_full['Delivery Delay (Days)'] = (sales_full ['Promised Delivery Date'] - sales_full ['Date']).dt.days

# Analyze average delivery delay by region
avg_delivery_delay_by_region = sales_full.groupby('Region')['Delivery Delay (Days)'].mean().sort_values()

# Display results
sales_by_region.head(), sales_by_product_group.head(), sales_by_product_type.head(), avg_delivery_delay_by_region.head()

avg_delivery_delay_by_region.plot(kind="bar", figsize=(10,6), color='lightblue')
plt.title('Average Delivery Delay by Region (in Days)')
plt.xlabel('Region')
plt.ylabel('Average Delivery Delay (Days)')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```
